```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('dataset/dataCovidIndonesia.csv')
data = data.loc[0:49, 'Kasus harian']
data = np.genfromtxt(data, dtype= None)
data
```

```python
def generasi_awal(pop_size, gene_per_pop):
    population_size = (pop_size, gene_per_pop)
    gene = np.random.uniform(low=0.1, high=0.9, size=population_size)
    return gene
```

```python
def individu_awal(pop_size, gene_per_pop):
    day_predict = []
    for i in range(gene_per_pop, pop_size):
        row = []
        for j in range(0, gene_per_pop):
            row.append(data[i])
            i=i-1
        day_predict.append(row)
    return day_predict
```

```python
def y_pred(gene, day_predict, gene_per_pop):
    y_pred_indv = []
    for i in range (0,gene_per_pop):
        row = []
        for j in gene:
            total = 0
            z = 0
            for k in j:
                if k<gene_per_pop-1:
                    g = k*day_predict[i][z]
                else:
                    g = k
                z=z+1
                total = total + g
            row.append(total)
        y_pred_indv.append(row)
    return y_pred_indv
```

```python
pop_size = 50
gene_per_pop = 3
generasiAwal = generasi_awal(pop_size, gene_per_pop)
generasiAwal
```

```python
day_predict = individu_awal(pop_size, gene_per_pop)
day_predict
```

```python
yPred = y_pred(generasiAwal, day_predict, gene_per_pop)
yPred
```
